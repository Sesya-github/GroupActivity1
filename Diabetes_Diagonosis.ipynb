```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('diabetes_model.csv')
```

```python
data
```

# 1). Count plot of Outcome

```python
sns.countplot(data['Outcome'])
```

# 2). Effect of Age on Outcome using Boxplot

```python
sns.boxplot(x='Outcome',y='Age',data=data)
```

# 3). Chances of getting diabetes, using Diabetes Pedigree Function

```python
sns.violinplot(x='Outcome',y='DiabetesPedigreeFunction',data=data)
```

# 4). Relationship between Insulin and Glucose

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='Glucose',y='Insulin',hue='Outcome',data=data)
```

# 5). Heatmap on correlation of features in the dataset

```python
plt.figure(figsize=(10,8))
sns.heatmap(data.corr(),annot=True)
plt.xticks(rotation=90)
```
